```python
from yahooquery import Screener
import pandas as pd
```

```python
s=Screener()
```

```python
s.available_screeners
```

```python
screen=s.get_screeners('top_mutual_funds_us', count=200)
print(screen)
```

```python
# Extract the tickers and company names
tickers = [item['symbol'] for item in screen['top_mutual_funds_us']['quotes']]
#names = [item['name'] for item in screen['top_mutual_funds_us']['quotes']]

# Convert to DataFrame
tickers_df = pd.DataFrame({'Ticker': tickers})

# Save to CSV
#tickers_df.to_csv('mutualfunds_tickers.csv', index=False)

print(tickers_df.head())
```

```python
import yfinance as yf

# List of tickers
tickers = tickers_df['Ticker'].tolist()

# Fetch company names
company_names = []
for ticker in tickers:
    stock = yf.Ticker(ticker)
    info = stock.info
    company_names.append(info.get('longName', 'N/A'))

# Convert to DataFrame
tickers_df = pd.DataFrame({'Ticker': tickers, 'Name': company_names})

# Save to CSV
tickers_df.to_csv('tickers_and_names.csv', index=False)

print(tickers_df.head())
```
