```python
import pdfplumber
import pandas as pd
from PIL import Image
```

```python
pdf = pdfplumber.open("abrdn SICAV I _ PROSPECTUS EN _ DECEMBER 2023.pdf")

#Load page_0
p144 = pdf.pages[143]

# Visualize the actual content on page_0
#Image.open("page_144.png")
```

```python
table = p144.extract_table()

df = pd.DataFrame(table[0:], columns=table[0]).dropna(how='all')
df = df.fillna(value=0)
df.drop(0)
df.dropna()

#make 0 blank
df = df.replace(0, '')

#IF the first column is empty, REMOVE THE ROW
df = df[df[df.columns[0]] != '']

#REMOVE FIRST ROW
df = df.iloc[1:]

#remove third and fourth column
#df = df.drop(df.columns[[2, 3]], axis=1)

df.info()
df




```

```python
#save df in excel file
df.to_excel("output.xlsx", index=False)
```
